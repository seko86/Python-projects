```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# I am loading 2 datasets
athletes =  pd. read_csv('C:/Users/Sebastian/Desktop/OLympics analysis/athlete_events.csv')
regions  =  pd. read_csv('C:/Users/Sebastian/Desktop/OLympics analysis/noc_regions.csv')
```

```python
# I am looking at first 10 rows of athletes dataset
athletes.head(10)
```

```python
# I am looking at first 10 rows of athletes dataset
regions.head(10)
```

```python
# I am joining two datasets using merge method and printing 5 first rows to see results
df = pd.merge(athletes, regions, how='left', on='NOC')
df.head()
```

```python
# Quick look at the total nr of rows and columns in my DataFrame
df.shape
```

```python
# Now I want to change last 2 column names to start with capital letters using rename method
df.rename(columns={'region':'Region', 'notes':'Notes'}, inplace=True)
```

```python
# Quick look if changes were made correctly
df.head()
```

```python
# Here I check data types for particular columns and check if null values appears in dataset
# I have some columns with null values to handle later
df.info()
```

```python
# Quick look at the main statistics based on numerical columns
# Interesting is fact that min age of athlete is 10 and max 97. Dataset includes all olympic games from 1896 till 2016.
df.describe()
```

```python
# Quick look at the main statistics based on text columns (strings)
# Interesting is fact that most frequent athlete joined olympics 58 times. Most games were held at 2000 olympics, 
# most popular city for hosting olympics is London, and most popular sport is football
df.describe(include='object')
```

```python
# Here I check particular numbers of rows with null values that I observed earlier
# Most null values I can observe in notes and medal columns
nan_values = df.isna().sum()
nan_values[nan_values > 0]
```

```python
# I would like now to filter dataset to see 5 first Polish athletes at Rio Olympics under 20
filt = (df['Team']=="Poland") & (df['Year'] == 2016) & (df['Age'] < 20)
df[filt].head()
```

```python
# I would like now to filter dataset to see Polish athletes at Rio Olympics with gold medals using query method
df.query('Team=="Poland" & Year== 2016 & Medal=="Gold"').head()
```

```python
# I want to see 10 top most participating athletes countries in Olympics
top_10_countries = df['Team'].value_counts().head(10).sort_values(ascending=False)
```

```python
# I want to present these countries and values on a barplot
plt.figure(figsize=(12,6))
plt.title('Overall Participation by Country')
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette='Set3')
plt.show()
```

```python
# I want to present histogram with athletes distribution by age
plt.figure(figsize=(12,6))
plt.title("Age distribution of the athletes")
plt.xlabel('Age')
plt.ylabel('Number of athletes')
plt.hist(df.Age, bins=np.arange(10,80,2), color = 'green', edgecolor='white')
plt.show()
```

```python
# I want now to compare summer and winter sport disciplines
summer_sports = df[df.Season == 'Summer']['Sport'].unique()
summer_sports
```

```python
winter_sports = df[df.Season == 'Winter']['Sport'].unique()
winter_sports
```

```python
# I want to see gender distribution on Olympics
gender_counts = df['Sex'].value_counts()
gender_counts
```

```python
# I want to present it on a pie chart
# Most of athletes are men with more than 70% share
plt.figure(figsize=(12,6))
plt.title("Gender distribution")
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=180)
plt.show()
```

```python
# Now I want to group data by year and see total numbers of female athletes at summer Olympics
# Looking at last 5 Olympics we can see growing numbers of femals
females_summer = df[(df['Sex'] == 'F') & (df['Season']=='Summer')][['Sex', 'Year']]
females_summer.groupby('Year').count().tail()
```

```python
# Let's have a look at the count plot
# In 2016 we had more than 6 thousands whereas at 1900 Olympics only 33 female participanted
sns.set(style="darkgrid")
plt.figure(figsize=(14,8))
sns.countplot(x='Year', data=females_summer, palette='Spectral')
plt.title('Women participation')
```

```python
# Gold medals athletes
gold_medals = df[df.Medal=='Gold']
gold_medals.head()
```

```python
# I want to see if column Age has null-values for Gold medalist
gold_medals[gold_medals['Age'].isna()].head()
```

```python
# I have null values so I want to delete them and see if I made it correctly
gold_medals = gold_medals[np.isfinite(gold_medals['Age'])]
gold_medals[gold_medals['Age'].isna()].head()
```

```python
# I want to see athletes that won gold medals with more than 60 years
# Only 6 such athletes won gold medals 
gold_medals[gold_medals['Age'] > 60]
```

```python
# Now let's have a look at sports where such athletes won gold medals
sport_60_gold = gold_medals['Sport'][gold_medals['Age'] > 60]
sport_60_gold
```

```python
# Let's present results in a countplot
sns.set(style="darkgrid")
plt.figure(figsize=(10,6))
sns.countplot(x=sport_60_gold, palette='Spectral')
plt.title('Disciplines with gold medal athletes above 60''s')
plt.show()
```

```python
# I want to present now gold medals by 5 top countries
gold_countries = gold_medals.Region.value_counts().reset_index(name='Gold medals').head()
gold_countries
```

```python
# And present it on a bar chart
golds = sns.catplot(x='index', y='Gold medals', data=gold_countries, height=5, kind='bar', palette='rocket')
golds.set_xlabels("Top 5 countries")
golds.set_ylabels("Nr of gold medals")
plt.title("Gold medals per country")
plt.show()
```

```python
# I want now present 5 most gold medals countries in Rio Olympics
max_year = df.Year.max()
print(max_year)
teams_gold = df[(df['Year'] == max_year) & (df['Medal'] == 'Gold')]
teams_gold.Region.value_counts().head(5)
```

```python
# Let's see 20 most gold medal countries on a horizontal bar plot
sns.barplot(x=teams_gold.Region.value_counts().head(20), y = teams_gold.Region.value_counts().head(20).index)
plt.ylabel(None)
plt.xlabel('Nr of gold medals at Rio Olympics')
plt.show()
```

```python
# At last I would like to see medalists with not null height and weight values
Medals = ['Bronze', 'Silver', 'Gold']
filt = (df['Height'].notnull()) & (df['Weight'].notnull() & df['Medal'].isin(Medals))
df[filt].head()
```

```python
# And present it on a scatter plot
plt.figure(figsize=(10,6))
axis = sns.scatterplot(x='Height', y='Weight', data=df[filt], hue='Sex')
plt.title('Medalists')
plt.show()
```
